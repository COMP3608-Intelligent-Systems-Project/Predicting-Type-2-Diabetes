```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Outcome', data=df, palette='Blues')
plt.title('Class Distribution - Pima Indians Dataset')
plt.xticks([0, 1], ['No Diabetes (0)', 'Diabetes (1)'])
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('pima_class_distribution.png')
plt.show()

print("Plot saved.")
```
