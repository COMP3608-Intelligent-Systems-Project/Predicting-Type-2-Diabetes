```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diabetes.csv')

print("Shape: " + str(df.shape))
print("\nFirst 5 rows:")
print(df.head())
```

```python
print("\nData types:")
print(df.dtypes)

print("\nBasic statistics:")
print(df.describe().round(2))
```

```python
print("Class distribution:")
print(df['Outcome'].value_counts())
print(f"\nPositive cases (diabetes): {(df['Outcome']==1).sum()} ({(df['Outcome']==1).mean()*100:.1f}%)")
print(f"Negative cases (no diabetes): {(df['Outcome']==0).sum()} ({(df['Outcome']==0).mean()*100:.1f}%)")
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Outcome', data=df, palette='Blues')
plt.title('Class Distribution - Pima Indians Dataset')
plt.xticks([0, 1], ['No Diabetes (0)', 'Diabetes (1)'])
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('pima_class_distribution.png')
plt.show()

print("Plot saved.")
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap - Pima Indians Dataset')
plt.tight_layout()
plt.savefig('pima_correlation_heatmap.png')
plt.show()

print("Plot saved.")
```

```python
df.hist(figsize=(12, 10), bins=20, color='steelblue', edgecolor='black')
plt.suptitle('Feature Distributions - Pima Indians Dataset', y=1.02, fontsize=14)
plt.tight_layout()
plt.savefig('pima_feature_distributions.png')
plt.show()

print("Plot saved.")
```
